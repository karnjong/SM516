# Line Graphs

First let's just create a random array of 10 elements as sample data:

```python
np.random.seed(0)

x = np.arange(10)
y = np.random.randint(0, 10, x.size)
```

Then to draw the graph, first create a plot with `subplots` then use `plot`:

```python
# Create a figure and axes
figure, ax = plt.subplots()

# Line of x vs y
ax.plot(x, y)
```

We can change the size of the plot by passing `figsize` into `subplots`:

```python
# Create a figure of size 8x8
figure, ax = plt.subplots(figsize=(8,8))
ax.plot(x, y)
```

There are a number of styling options, for instance to make a dashed red line:

```python
figure, ax = plt.subplots()

# `r` means read, `--` means dashed
ax.plot(x, y, 'r--')
```

You can add a title and axes labels:

```python
figure, ax = plt.subplots()
ax.plot(x, y)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Y v X')
```

# Scatter Plots

Similarly, you can create scatter plots:

```python
figure, ax = plt.subplots()
ax.scatter(x, y)
```

Change the size of the dots with the keyword arg `s`:

```python
figure, ax = plt.subplots()

# Make each point size 100
ax.scatter(x, y, s=100)
```

`s` can also be a list with a size corresponding to each point.  For example, to scale the point size based on the y value:

```python
figure, ax = plt.subplots()

# Make each point size y value*20
ax.scatter(x, y, s=y*20)
```

# Combining Plots + Legends

You can put multiple plots onto the same axes.  For example, combining the scatter plot with the line plot:

```python
figure, ax = plt.subplots()

ax.plot(x, y)
ax.scatter(x, y, s=100)
```

With multiple plots, a legend is useful.  Include a `label` parameter when creating the plots and then call legend:

```python
figure, ax = plt.subplots()

ax.plot(x, y, label='line')
ax.scatter(x, y, s=100, label='points')

ax.legend()
```

The legend location can be changed with `loc`:

```python
figure, ax = plt.subplots()

ax.plot(x, y, label='line')
ax.scatter(x, y, s=100, label='points')

ax.legend(loc='upper left')
```

The options for `loc` can either be a descriptive string or an integer code.  The strings and corresponding codes are:

    'best'         : 0,
    'upper right'  : 1,
    'upper left'   : 2,
    'lower left'   : 3,
    'lower right'  : 4,
    'right'        : 5,
    'center left'  : 6,
    'center right' : 7,
    'lower center' : 8,
    'upper center' : 9,
    'center'       : 10,
    
You can also crate plots side-by-side in the same figure.  In the call to `subplots` use `nrows` and `ncols` to specify how many axes you want vertically and horizontally.  This will then return an NxN tuple of axes.  For example, to create two 6x6 plots next to each other:

```python
nrows = 1 # Just one row
ncols = 2 # of two plots
figsize = (12,6) # We want each plot to be 6x6, so the width should be 6*2=12

# The axes are returned as a 1-tuple with a 2-tuple of the axes
figure, ((ax0, ax1)) = plt.subplots(figsize=figsize, nrows=nrows, ncols=ncols)

ax0.plot(x, y)
ax1.scatter(x, y, s=100)
```

To instead put one on top of the other:

```python
nrows = 2 # This time we want 2 rows
ncols = 1 # A just one column
figsize = (6,12) # We need to switch the figsize to be height 12

# Notice we also need to modify the tuple to be a 2-tuple of 1-tuples
figure, ((ax0), (ax1)) = plt.subplots(figsize=figsize, nrows=nrows, ncols=ncols)

ax0.plot(x, y)
ax1.scatter(x, y, s=100)
```

# Histograms

To create a histogram, use `hist`:

```python
# Let's craete some normally distributed data:
n = np.random.randn(1000)

figure, ax = plt.subplots()
ax.hist(n)
```

By default it uses 10 bins, but that can be changed:

```python
figure, ax = plt.subplots()

# Use 20 bins instead
ax.hist(n, bins=20)
```

# Other Styling Options

There are a number of style sheets included:

```python
plt.style.available
```

For example, ggplot:

```python
plt.style.use('ggplot')

figure, ax = plt.subplots()
ax.plot(x, y)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Y v X')
```
